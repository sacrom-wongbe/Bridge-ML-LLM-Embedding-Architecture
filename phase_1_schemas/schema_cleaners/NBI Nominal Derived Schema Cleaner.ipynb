```python
import json
import os

# Load all three schemas
numerical_path = os.path.join(os.getcwd(), "schemas", "nbi_numerical_schema.json")
numerical_coded_path = os.path.join(os.getcwd(), "schemas", "nbi_numerical_coded_schema.json")
nominal_derived_path = os.path.join(os.getcwd(), "schemas", "nbi_nominal_derived_schema.json")

with open(numerical_path, "r", encoding="utf-8") as f:
    numerical_schema = json.load(f)

with open(numerical_coded_path, "r", encoding="utf-8") as f:
    numerical_coded_schema = json.load(f)

with open(nominal_derived_path, "r", encoding="utf-8") as f:
    nominal_derived_schema = json.load(f)

# Create a combined mapping of source column -> description
source_descriptions = {}
for col, meta in numerical_schema.items():
    if isinstance(meta, dict) and "description" in meta:
        source_descriptions[col] = meta["description"]

for col, meta in numerical_coded_schema.items():
    if isinstance(meta, dict) and "description" in meta:
        source_descriptions[col] = meta["description"]

# Update nominal_derived entries with descriptions from their sources
updated = 0
missing_sources = []

for col, meta in nominal_derived_schema.items():
    if isinstance(meta, dict) and "source" in meta:
        source = meta["source"]
        if source in source_descriptions:
            meta["description"] = source_descriptions[source]
            updated += 1
        else:
            missing_sources.append(f"{col} -> {source}")

# Save updated schema
with open(nominal_derived_path, "w", encoding="utf-8") as f:
    json.dump(nominal_derived_schema, f, indent=2, ensure_ascii=False)

print(f"✅ Updated {updated} nominal_derived columns with descriptions from their sources")
if missing_sources:
    print(f"⚠️ {len(missing_sources)} columns had sources not found:")
    for item in missing_sources:
        print(f"  - {item}")
print(f"Saved to {nominal_derived_path}")
```
